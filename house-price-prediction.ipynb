```python
#dropping columns Alley,PoolQC,Fence,MiscFeature,MasVnrType,FireplaceQu
# as they have max number of nulls which is >40% data.

data_new = data.drop(columns = ['LotFrontage','Alley', 'PoolQC' ,'Fence' , 'MiscFeature', 'MasVnrType', 'FireplaceQu'])
```

```python
data_new
```

```python
data_null2 = data_new.isnull().sum(axis=0)
data_null3 = data_null2[data_null2!=0]
data_null3
```

```python
#we can drop these rows as they are very less count of data missing.

data_new.dropna(inplace = True)
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

for i in range(0,data_new.shape[1]):
    if data_new.dtypes[i] == 'object':
        data_new[data_new.columns[i]] = encoder.fit_transform(data_new[data_new.columns[i]])
```

```python
data_new
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
x = data_new.drop(columns = ['SalePrice'])
y = data_new[['SalePrice']]
```

```python
x
```

```python
y
```

```python
train_x, val_x, train_y, val_y = train_test_split(x,y,random_state=1)
```

```python
regressor = RandomForestRegressor()
regressor.fit(train_x,train_y)
```

```python
predictions = regressor.predict(val_x)
```

```python
predictions
```

```python
val_mae = mean_absolute_error(predictions, val_y)
```

```python
val_mae
```

```python
predictions = pd.DataFrame(predictions)
```

```python
predictions.to_csv("house_price_predict_submission.csv", index= False)
```
